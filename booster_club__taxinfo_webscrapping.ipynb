```
testurl=requests.get('https://apps.irs.gov/app/eos/searchFromCORResults.do?resultsPerPage=250&searchChoice=&names=booster&city=&state=All+States&country=US&ein1=&zipCode=&dispatchMethod=searchCopyOfReturns&&&indexOfFirstRow=250&sortColumn=orgCountry&isDescending=false&selectedIndex=0')
```

```
testurl.text
```

```
import re
```

```
import bs4
soup = bs4.BeautifulSoup(testurl.text,"html5lib")
links = soup.find_all(href=re.compile("^/app/eos/display"))
linktable=[str(i) for i in links]
```

```
len(linktable)
```

```
namelist=[]
for element in linktable:
  p=re.compile('>.*</a')
  m=str(re.findall(p,element))
  m=m.strip('[]')
  m=m.strip('\'')
  m=m.strip("\'")
  m=m.replace('</a','')
  m=m.replace('>','')
  namelist.append(m)
```

```
namelist
```

```
newtable=[]
for item in linktable:
  item=item.replace('<a href=','')
  item=item.replace('</a>','')
  item=item.replace('"','')
  item=item.replace('amp;','')
  newtable.append(str(item))
  
```

```
newtable[3]
```

```

for i,url in enumerate(newtable):
  newtable[i]="https://apps.irs.gov"+url
  p=re.compile(r">.*")
  newtable[i]=str(re.sub(p,'',newtable[i]))
```

```
len(newtable)
```

```
newtable[3]
```

```
timelist=list()
taxtypelist=list()
for i,url in enumerate(newtable):
  test=requests.get(newtable[i])
  soup1 = bs4.BeautifulSoup(test.text,"html5lib")
  testlinks = soup1.find_all(href=re.compile("^/pub/epostcard/cor"))
  testlinks=testlinks[0]

  
  testtable=[]
  for item in testlinks:
    item=item.replace('<a href=','')
    item=item.replace('</a>','')
    item=item.replace('"','')
    item=item.replace("Form","")
    item=item.replace("Filing","")
    item=item.rstrip()
    newlist=item.split(" ")
    
    time=newlist[0][0:4]
    timelist.append(time)
    taxtype=newlist[0][6:]
    taxtypelist.append(taxtype)

  
```

```
timelist
```

```
taxtypelist
```

```
import pandas as pd
df=pd.DataFrame({"link":newtable,"name":namelist,"time":timelist,"taxtype":taxtypelist})
```

```
df
```

```
df.to_csv(r'booster_final-2.csv',index=False,sep=',')
```
